```python
import random
a = random.sample(range(1,10),4)
count = 1
b = input("請你輸入四個數字(不重複)")
c = list(str(b))
d = [int(i) for i in c]
while d != a:
  count_a = 0
  count_b = 0
  if count != 1 and len(d) == 4:
    b = input("請你輸入四個數字(不重複)")
    c = list(str(b))
    d = [int(i) for i in c]
    if d == a:
      print(b+':4A0B')
      print('答對了! 猜題次數:'+str(count)+'次')
      break
  elif count != 1 and len(d) > 4:
    b = input("輸入數字過多, 請你輸入四個數字(不重複)")
    c = list(str(b))
    d = [int(i) for i in c]
  elif count != 1 and len(d) < 4:
    b = input("輸入數字過少, 請你輸入四個數字(不重複)")
    c = list(str(b))
    d = [int(i) for i in c]
  if len(d) == 4:
    count += 1
    for m in range(len(d)):
      for n in range(len(d)):
        if d[m] == a[n] and m==n:
          count_a += 1
        elif d[m] == a[n] and m!=n:
          count_b += 1
    print(b+":"+str(count_a)+'A'+str(count_b)+"B")
if d == a and count == 1:
  print(b+':4A0B')
  print('答對了! 猜題次數:1')
```
